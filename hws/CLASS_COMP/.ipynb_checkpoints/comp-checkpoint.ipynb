## Модель 1. Линейная регрессия,
### или почему это неудачная модель на бинарной классификации.

Рассмотрим модель линейной регрессии. Она как правило плохо подходит для решения задач бинарной классификации(все же она хорошо подходить для аппроксимации непрерывных функций), а значит послужить неплохой отправной точкой для классификации.

```python
from sklearn.linear_model import LinearRegression
```

```python
instinct = LinearRegression().fit(X, y)
```

```python
instinct.score(X, y)
```

Добавили удобную функцию для подсчета `accuracy` моделей:

```python
count = sum((instinct.predict(X) >= 0.5).astype(int) == y)
count / 1000
```

```python
def acc(model, X, y):
    return sum((model.predict(X) >= 0.5).astype(int) == y) / X.shape[0]
```

```python
acc(instinct, X, y)
```

Точность предсказания получилась заметно выше монетки, так что можно пробовать разновидности регрессии:

```python
from sklearn.linear_model import Ridge, Lasso
```

Так как лучшего способа угадать альфу нет, выберем те, что дают наибольший `score`

```python
rscores, lscores = [], []
for n in range(1000):
    ridge = Ridge(alpha=10.0 * (n + 1) / 1000, fit_intercept=False)
    ridge.fit(X, y)
    lasso = Lasso(alpha=10.0 * (n + 1) / 1000, fit_intercept=False)
    lasso.fit(X, y)
    rscores.append(ridge.score(X, y))
    lscores.append(lasso.score(X, y))
```

```python
np.max(rscores), np.max(lscores)
```

Для читаемости строчки были удалены, но наибольший `rscore` достигался при `alpha=0.01`, их точности:

```python
ridge = Ridge(alpha=0.01, fit_intercept=False)
lasso = Lasso(alpha=0.01, fit_intercept=False)
ridge.fit(X, y)
lasso.fit(X, y)
acc(ridge, X, y), acc(lasso, X, y)
```

В какой то момент была идея усреднять, или усреднять с весами модели, брать голосовашку (брать максимум вероятностей предсказания) в голову во время соревнования не приходило. 

Тем не менее, для этого нужно было реализовать усредняющие модели:

```python
class average_model:
    def __init__(self, models):
        self.models = models
    def fit(self, X, y):
        for model in self.models:
            model.fit(X, y)
    def predict(self, X):
        return sum([model.predict(X) for model in self.models]) / len(self.models)
```

```python
class weighted_model:
    def __init__(self, models, weights):
        self.models = models
        self.weights = np.array(weights) / sum(weights)
    def fit(self, X, y):
        for model in self.models:
            model.fit(X, y)
    def predict(self, X):
        return sum([model.predict(X) * self.weights[i] for i, model in enumerate(self.models)]) / len(self.models)
```

Базовый тест:

```python

acheck = average_model([LinearRegression()])
acheck.fit(X, y)
acc(acheck, X, y)
```

Далее была идея усреднять три нынешние модели, в надежде того, что какие то из них будут гораздо увереннее в своих предсказаниях ежели другие модели, тем самым увеличивая `accuracy`. Естественно голосовашка справляется с задачей заметно лучше, но с чуть большей ошибкой второго рода(или первого, смотря что мы считаем нулевыми гипотезами)

```python
averagem = average_model([LinearRegression(), Ridge(alpha=0.01, fit_intercept=False),
                  Lasso(alpha=0.01, fit_intercept=False)])
averagem.fit(X, y)
acc(averagem, X, y)
```

```python
weightm = weighted_model([LinearRegression(), Ridge(alpha=0.01, fit_intercept=False),
                  Lasso(alpha=0.01, fit_intercept=False)], [0.735, 0.689, 0.662])
weightm.fit(X, y)
acc(weightm, X, y)
```

Как видно, ни одно из усреднений не улучшило модель. Пора засылать лучшую из версий(линейная регрессия) и придумывать новые модели

## Модель 2. Решающее дерево,

Для начала попробуем модель решающего дерева.

Сделаем удивленное лицо узнав, что это не 100 и даже не 90 процентов. Стоит отметить, однако, что бенчмарк на паблике мы прошли, но лишь едва едва. Дает повод подозревать, что на привате бенчмарк подобное решение не пройдет.

##  Модель 3. ближайшие соседи,
### или как оценивать (не)идеальные модели точнее.

 Пока отложим модель в сторону и попробуем другую:

```python
from sklearn.neighbors import KNeighborsClassifier
```

Попробуем разное число соседей. Как видно, точность убывает с ростом соседей, причем точность 100% достигается, при одном соседе.

```python
for n in range(1, 20):
    KNN = KNeighborsClassifier(n_neighbors=n)
    KNN.fit(X, y)
    print(acc(KNN, X, y))
```

Попробуем усреднить модель с предыдущей. Действительно, две идейно разные модели смогли дать полностью точный результат, вероятно как то смешав их можно достигнуть лучшего результата.

```python
coolaverage = average_model([KNeighborsClassifier(), DecisionTreeClassifier()])
coolaverage.fit(X, y)
acc(coolaverage, X, y)
```

Однако отнюдь, придется опять поплакать об упущенной возможности реализовать голосовашку.

```python
y2 = coolaverage.predict(X_test)
y2 = (y2 >= 0.5).astype(int)
y2
```

На всякий случай убедимся, что предсказание отличается от дерева.

```python
np.all(newY == y2)
```

Как далее будет показано, это не единственные две модели, способные давать идеальный `accuracy` на обучающей выборке. Таким образом, нам потребуется *кросс-валидация*:

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
print(cross_val_score(DecisionTreeClassifier(), X, y, cv=5).mean(),
cross_val_score(KNeighborsClassifier(n_neighbors=1), X, y, cv=5).mean())
```

Отсюда видно, что ближайшие соседи дают заметно худший результат, ежели решающее дерево. Зашлем усредненную модель и продолжим развивать дерево:

```python
ans2 = pd.DataFrame(data={'target': y2, 'Id': idd})
ans2.head()
```

```python
ans2.to_csv('out3.csv', index=False,
          compression=None)
```

## Модель 4. XGBoost,
### или улучшение дерева.

Мы видим, что дерево, даже с параметрами по умолчанию явлется очень перспективной моделью классификации, а значит можно попробовать использовать `XGBoost`

```python
import xgboost as xgb
```

У `XGBoost`, как известно, очень много параметров, их удобнее будет перебирать с помощью `GridSearch`. Времени много, так что возьмем (относительно) большие наборы параметров, вращаясь вокруг изначальных, и выберем лучшую модель.

```python
param_grid = [{'max_depth': [2, 4, 6, 8, 10], 'eta': [i / 10 for i in range(1, 11)], 'gamma': [0, 1, 2, 5, 10],
              'subsample': [0.5, 0.8, 1]}]
searcher = GridSearchCV(xgb.XGBRegressor(tree_method="exact"), param_grid, cv=5)
searcher.fit(X, y)
```

```python
best_one = searcher.best_estimator_
acc(best_one, X, y)
```

Как и ожидалось, accuracy 1.0. Будучи самоуверенным, я решил не смотреть даже на крос-вал и послал так.

```python
coolY = (best_one.predict(X_test) >= 0.5).astype(int)
```

```python
coolAns = pd.DataFrame(data={'target': coolY, 'Id': idd})
coolAns.head()
```

```python
coolAns.to_csv('cool_out.csv', index=False,
          compression=None)
```

## Вредные советы. 
### А что если заполнить пропуски с помощью KNN?

Название говорит само за себя. Я где то слышал, что `XGBoost` дает лучшие результаты при заполнении пропусков с помощью `KNN`. Возможно так и вышло бы, если бы я выбрал большее число соседей, однако я почему то выбрал то, которое лучше всего обучает модель классификации, в которой пропуски заполнены средним.

```python
from sklearn.impute import KNNImputer
imputer = imputer = KNNImputer(n_neighbors=1, weights="uniform")
```

```python
y, X = df['target'], df[df.columns[1:]]
y, X
```

```python
X = imputer.fit_transform(X)
X = pd.DataFrame(X, columns = ['feature' + str(i + 1) for i in range(13)])
X.head()
```

Сверху было заполнение пропусков.

```python
param_grid = [{'max_depth': [2, 4, 6, 8, 10], 'eta': [i / 10 for i in range(1, 11)], 'gamma': [0, 1, 2, 5, 10],
              'subsample': [0.5, 0.8, 1]}]
searcher = GridSearchCV(xgb.XGBRegressor(tree_method="exact"), param_grid, cv=5)
searcher.fit(X, y)
```

```python
best_one = searcher.best_estimator_
acc(best_one, X, y)
```

Наблюдая все еще полный `accuracy` я решил заслать эту версию(стоило рассмотреть крос-вал)

```python
X_test = pd.read_csv('test.csv')
X_test.head()
```

```python
X_test = imputer.fit_transform(X_test)
X_test = pd.DataFrame(X_test, columns = ['feature' + str(i + 1) for i in range(13)] + ['Id'])
idd = X_test['Id']
del X_test['Id']
X_test.head()
```

Естественно, чтобы не терять точности, пропуски необходимо было заполнять тем же методом.

```python
coolerY = (best_one.predict(X_test) >= 0.5).astype(int)
```

```python
idd = [int(t) for t in idd]
```

```python
coolAns = pd.DataFrame(data={'target': coolerY, 'Id': idd})
coolAns.head()
```
